# arima Model(Autoregressive integrated moving average)

# Using Train data Predicting Val Data Per Category

```python
def ARIMA_model(category_name):
    predictions=list()
    true_predictions=list()
    category_data=pd.DataFrame(train.groupby(['Productcat']).get_group(category_name))
    category_data=category_data.set_index('Date')
    category_data=category_data.reindex(pd.date_range(start=category_data.index[0], end=category_data.index[-1], freq='d'))
    category_data.index=category_data.index.normalize()
    category_data=category_data.drop(["Productcat"],axis=1)
    category_data=category_data.interpolate(method ='linear', limit_direction ='backward', limit = 3)
    category_data.dropna(inplace=True)
    category_val=pd.DataFrame(val.groupby(['Productcat']).get_group(category_name))
    category_val=category_val.set_index('Date')
    category_val=category_val.drop("Productcat",axis=1)
    category_val=category_val.reindex(pd.date_range(start=category_val.index[0], end=category_val.index[-1], freq='d'))
    category_val.index=category_val.index.normalize()
    category_val=category_val.interpolate(method ='linear', limit_direction ='backward', limit = 3)
    category_val.dropna(inplace=True)
    category_val_index=category_val.index
    category_data=category_data.values
    category_val=category_val.values
    history=[x for x in category_data]
    test=[y for y in category_val]
    for i in range(len(test)):
        if category_name not in ["ProdCat4","ProdCat6","ProdCat63","ProdCat64","ProdCat65","ProdCat66",'ProdCat132','ProdCat142','ProdCat187','ProdCat192','ProdCat201']:
            model = ARIMA(history, order=(5,1,0))
        else:
            model = ARIMA(history, order=(0,0,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        true_predictions.append(test[i])
        history.append(test[i])
    error = mean_absolute_error(true_predictions, predictions)
    print('Val MAE: %.3f' % error)
    error = mean_squared_error(true_predictions, predictions)
    print('Val MSE: %.3f' % error)
    error = sqrt(mean_squared_error(true_predictions, predictions))
    print('Val RMSE: %.3f' % error)
    return predictions,true_predictions,category_val_index,history
```

# Using Train data and Val Data Per Category to predict testing Data

```python
def test_data_predict(name,history):
    predictions_test=list()
    category_test=pd.DataFrame(train.groupby(['Productcat']).get_group(name))
    category_test=category_test.set_index('Date')
    #category_data=category_data.reindex(pd.date_range(start=category_data.index[0], end=category_data.index[-1], freq='d'))
    category_test.index=category_test.index.normalize()
    category_test=category_test.drop(["Productcat"],axis=1)
    #category_data=category_data.interpolate(method ='linear', limit_direction ='backward', limit = 2)
    category_test_index=category_test.index
    category_test=category_test.values
    #category_test_index=category_test.index
    test_data=[x for x in category_test]
    for j in range(len(test_data)):
        if name not in ["ProdCat4","ProdCat6","ProdCat63","ProdCat64","ProdCat65","ProdCat66","ProdCat132","ProdCat142","ProdCat187","ProdCat192","ProdCat201"]:
            model = ARIMA(history, order=(5,1,0))
        else:
            model = ARIMA(history, order=(1,0,1))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions_test.append(yhat)
        history.append(yhat)
    return predictions_test,category_test_index
```
